```python
df_train = pd.read_csv('/content/drive/MyDrive/black-friday sales/train.csv')
df_train.head()
```

```python

df_test = pd.read_csv('/content/drive/MyDrive/black-friday sales/test.csv')
df_test.head()
```

```python
#Merge both train n test
df = df_train.append(df_test)
df.head()
```

```python
# Specify custom directory path
custom_directory = "/content/drive/MyDrive/black-friday sales"

# Check if the directory exists, if not, create it
if not os.path.exists(custom_directory):
    os.makedirs(custom_directory)

# Save the combined DataFrame to a CSV file in the custom directory
csv_file = os.path.join(custom_directory, "combined_data.csv")
df.to_csv(csv_file, index=False)

print("Combined dataset has been saved to CSV successfully at custom directory:", custom_directory)
```
